# **Analysis international football result from 1872 to 2026**

Este conjunto de datos incluye **49.016** resultados de partidos internacionales de fútbol desde el primer partido oficial en 1872 hasta 2025. Los partidos van desde la Copa Mundial de la FIFA hasta la FIFI Wild Cup y partidos amistosos regulares. Los partidos son estrictamente internacionales masculinos completos y los datos no incluyen los Juegos Olímpicos ni partidos en los que al menos uno de los equipos fue el equipo B nacional, sub-23 o un equipo selecto de liga.

### **results.csv** :

- **date** - Fecha del partido

- **home_team** - el nombre del equipo local

- **away_team** - el nombre del equipo visitante

- **home_score** - Marcador del equipo local a tiempo reglamentario incluyendo prórroga, sin contar las tandas de penaltis

- **away_score** - Marcador del equipo visitante a tiempo reglamentario incluyendo prórroga, sin contar las tandas de penaltis

- **tournament** - el nombre del torneo

- **city** - el nombre de la ciudad/pueblo/unidad administrativa donde se jugó el partido

- **country** - el nombre del país donde se jugó el partido

- **neutral** - Columna VERDADERO/FALSO que indica si el partido se jugó en un lugar neutral

### **shootouts.csv** :

- **date** - Fecha del partido

- **home_team** - el nombre del equipo local

- **away_team** - el nombre del equipo visitante

- **winner** - ganador de la tanda de penaltis

- **first_shooter** - el equipo que salió primero en la tanda de penaltis

### **goalscorers.csv** :

- **date** - Fecha del partido

- **home_team** - el nombre del equipo local

- **away_team** - el nombre del equipo visitante

- **team** - nombre del equipo que marcó el gol

- **scorer** - nombre del jugador que marcó el gol

- **own_goal** - si el gol fue un gol en propia puerta

- **penalty** - si el gol fue penalti


### **former_names.csv** :

- **current** - nombre del equipo tal como se usa actualmente (o el apellido si el equipo ya no existe)

- **former** - nombre anterior utilizado por dicho equipo

- **start_date** - fecha de inicio de cuando se usó el nombre anterior

- **end_date** - fecha de finalización de cuando se usó el nombre anterior

### **Limpieza de datos (Pandas: dropna, fillna, replace, etc.)**

Validación de nulos del dataframe results

````python
results.isnull().sum()
````

````python
results_null_scores = results[
    results['home_score'].isna() | results['away_score'].isna()
]
results_null_scores
````

Eliminar nulos del dataframe results

````python
results_f = results.dropna(subset=['home_score', 'away_score']).copy()

results_f.isnull().sum()
````

Cambiar el formato de fecha del dataframe results

````python
# Convertir la columna date de forma segura; valores inválidos -> NaT
results_f.loc[:, 'date'] = pd.to_datetime(results_f['date'], errors='coerce')
````

````python
goals.isnull().sum()
````

````python
goals_null_scores = goals[
    goals['scorer'].isna() & goals['minute'].isna()
]
goals_null_scores
````

Eliminar nulos del dataframe goals

````python
# Crear copia explícita para evitar SettingWithCopyWarning
goals_f = goals.dropna(subset=['scorer', 'minute']).copy()
goals_f.isnull().sum()
````

Validación de valores negativos en 'minute' del dataframe 'goals_f':

````python
# Rellenar valores faltantes en 'minute' de forma segura
goals_f.loc[:, 'minute'] = goals_f['minute'].fillna(-1)
````
